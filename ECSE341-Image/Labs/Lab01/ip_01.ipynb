## **Part 1**

Reading using PIL

```
from PIL import Image   
im=Image.open(r"img.jpg")
width, height = im.size
print("width:",width)
print("height:",height)
im1 = im.save("pil.jpg")
im
```

Reading using Matplotlib

```
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
img = mpimg.imread('img.jpg')
plt.imshow(img)
plt.imsave("matplot.jpg",img)
print(img)
```

Read using skimage

```
from skimage import io
img=io.imread("img.jpg")
io.imshow(img)
print(img)
io.imsave("skimage.jpg",img)
```

## **Part 2**

```
import numpy as np

ROWS = 512    
COLS =  512

fin = open('lena_gray.raw')     
print(fin)

img = np.fromfile(fin, dtype = np.uint8, count = ROWS * COLS)
print("Dimension of the old image array: ", img.ndim)
print("Size of the old image array: ", img.size)

img.shape = (img.size // COLS, COLS)
print("Dimension of the NEW image array: ", img.ndim)
print("2D array: ",img)
print("The shape of the original image array is: ", img.shape)

img.astype('int8').tofile('NewImage.raw')
fin.close()
```
