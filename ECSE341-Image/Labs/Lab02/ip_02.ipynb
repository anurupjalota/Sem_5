B) RGB to greyscale by numpy n matplotlib

```
an_image = mpimg.imread("img.jpg")
plt.subplot(211)
plt.imshow(an_image)
rgb_weights = [0.2989, 0.5870, 0.1140]
grayscale_image = np.dot(an_image[...,:3], rgb_weights)
plt.subplot(212)
plt.imshow(grayscale_image, cmap=plt.get_cmap("gray"))
```

Explanation of part b: I first converted the image to a matrix of 3D array and then updated its rgb weights as 0.2989 for red, 0.5870 for green and 0.1140 for blue. But this is not the final output as it displays a bluish green image this is becuase the green weight has the largest weight. Also, our image is in 3D channel whereas greyscale images are in 1D channel hence I converted the image to 1D channel to get a true grayscale output.

Question 4: Image resizing

```
img = cv2.imread(r"img2.jpg", cv2.IMREAD_UNCHANGED)
 
print('Original Dimensions : ',img.shape)
 
scale_percent = 80 # percent of original size
width = int(img.shape[1] * scale_percent / 100)
height = int(img.shape[0] * scale_percent / 100)
dim = (width, height)

# resize image
#bilinear
resized_bi = cv2.resize(img, (500,500), interpolation = cv2.INTER_LINEAR)
#nearest neighbour
resized_nn = cv2.resize(img, (500,500), interpolation = cv2.INTER_NEAREST)
#bicubic
resized_bc = cv2.resize(img, (500,500), interpolation = cv2.INTER_CUBIC)

cv2_imshow(img)
cv2.waitKey(0)
cv2_imshow(resized_bi)
cv2.imwrite("Bilinear.jpg", resized_bi)
cv2.waitKey(0)
cv2_imshow(resized_nn)
cv2.imwrite("Nearest Neighbour.jpg", resized_nn)
cv2.waitKey(0)
cv2_imshow(resized_bc)
cv2.imwrite("Bicubic.jpg", resized_bc)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

1)	Compared to the original image the image with interpolation using nearest neighbour the image around the centre looked to be more distorted that is it appeared to me as if the pixels were overstretching vertically more as compared to horizontal stretched.

2) Compared to the original image the image with interpolation using bicubic appeared to be better quality as compared to nearest neighbour. 
Also out of 3 resizing this seems to be the most sharp. 

3)	Compared to the original image the image with interpolation using bilinear appeared to me the most blurred image. If the image was a bit sharp it would have be the best output amongst all.
