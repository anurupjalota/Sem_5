## **Working with Series**

Create two series

```
import numpy as np
se1=pd.Series(np.empty(100,dtype=int))
se2=pd.Series(np.empty(100,dtype=int))
```

Arithmetic Operations

```
add=se1.add(se2,fill_value=0)
sub=se1.sub(se2,fill_value=0)
mul=se1.mul(se2,fill_value=0)
div=se1.div(se2,fill_value=0)
```

Store all the 6 series to a dataframe

```
pd.DataFrame({"Series 1":se1, "Series 2":se2, "Add":add, 
              "Sub":sub, "Multiply":mul,"Divide":div})
```
